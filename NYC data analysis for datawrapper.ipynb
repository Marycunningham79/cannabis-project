```python
import pandas as pd
```

```python
df_raw = pd.read_csv("data/nyc-and-non-nyc.csv")
df_raw
```

```python
df_nyc = df_raw[df_raw["County"].eq("NYC Total")] 
df_nyc
```

```python
total_arrest = df_nyc.groupby(by="Race_Ethnicity").sum()
total_arrest
```
